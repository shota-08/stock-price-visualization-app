```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import altair as alt
```

```python
days = 100
tickers = {
    "SP500": "2558.T",       # 東証ETF - S&P500
    "AllCountry": "2559.T",  # 東証ETF - MSCIオールカントリー
    "Nikkei": "1321.T",      # 東証ETF - 日経平均
    "TOPIX": "1308.T",       # 東証ETF - TOPIX
    "FANG": "316A.T"         # 東証ETF - FANG+
}

df = pd.DataFrame()

for index in tickers.keys():
    tkr = yf.Ticker(tickers[index])
    hist = tkr.history(period=f'{days}d')
    hist.index = hist.index.strftime("%d %B %Y")
    hist = hist[["Close"]]
    hist.columns = [index]
    hist = hist.T
    hist.index.name = "Name"
    df = pd.concat([df, hist])
```

```python
index = ["SP500", "AllCountry"]
data = df.loc[index]
data
```

```python
data = data.T.reset_index()
data
```

```python
data = pd.melt(data, id_vars=["Date"]).rename(
    columns={"value":"Prices(JPN)"}
)
data.head()
```

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x="Date:T",
        y=alt.Y("Prices(JPN)", stack=None, scale=alt.Scale(domain=[20000, 30000])),
        color="Name:N"
    )
)

chart
```
