```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
days = 20
tickers = {
    "S&P500" : "2558.T",
    "AllCountry" : "2559.T",
    "Nikkei" : "1321.T",
    "TOPIX" : "1308.T",
    "FANG+" : "316A.T"
}
index = "S&P500"

tkr = yf.Ticker(tickers[index])
hist = tkr.history(period=f'{days}d')
hist.index = hist.index.strftime("%d %B %Y")
hist = hist[["Close"]]
hist.columns = [index]
hist = hist.T
hist.index.name = "Name"
hist
```

```python
days = 20
tickers = {
    "S&P500": "2558.T",      # 東証ETF - S&P500
    "AllCountry": "2559.T",  # 東証ETF - MSCIオールカントリー
    "Nikkei": "1321.T",      # 東証ETF - 日経平均
    "TOPIX": "1308.T",       # 東証ETF - TOPIX
    "FANG+": "316A.T"        # 東証ETF - FANG+
}

df = pd.DataFrame()

for index in tickers.keys():
    tkr = yf.Ticker(tickers[index])
    hist = tkr.history(period=f'{days}d')
    hist.index = hist.index.strftime("%d %B %Y")
    hist = hist[["Close"]]
    hist.columns = [index]
    hist = hist.T
    hist.index.name = "Name"
    df = pd.concat([df, hist])
```
